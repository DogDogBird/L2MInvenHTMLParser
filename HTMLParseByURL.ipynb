```python
def ParseL2MCollectionData(url):
    FullHTML = getFullHTMLFromURL(url)
    #print(FullHTML)

    splitHTML1 = []
    splitHTML1 = FullHTML.split("<th>컬렉션</th>")
    #print(splitHTML1[1])

    splitHTML2 = []
    splitHTML2 = splitHTML1[1].split("<div id = \"lineage2mRight\">")
    #print(splitHTML2[0])

    itemHTML = []
    itemHTML = splitHTML2[0].split("<td class=\"name\">")
    #print(itemHTML[1])

    craftItem = []
    temp1 = []
    for i in range(1,len(itemHTML)):
        #컬렉션 등급
        temp1 = []
        temp1 = itemHTML[i].split("<p class=\"name\" data-collection-grade=\"")
        collectionLevel = temp1[1].split("\">")[0]
        temp2 = temp1[1].split("\">")[1]

        # 컬렉션 이름
        collectionName = temp2.split("</p>")[0]
        collectionName = collectionName.strip()
        collectionName = collectionName.replace(","," ")
        #print(collectionName)

        # 컬렉션 옵션
        t_collectionOption = itemHTML[i].split("<p class=\"option\">")[1]
        collectionOption = t_collectionOption.split("</p>")[0]
        collectionOption = collectionOption.strip()
        collectionOptions = collectionOption.split(",")
        for j in range (len(collectionOptions)):
            collectionOptions[j] = collectionOptions[j].strip()

        collectionOptionNames = []
        collectionOptionValues = []
        for j in range (len(collectionOptions)):
            if(collectionOptions[j].find("치명타 저항") > -1):
                collectionOptionName = collectionOptions[j].split(" ")[0] + collectionOptions[j].split(" ")[1]
                collectionOptionNames.append(collectionOptionName)
                collectionOptionValue = collectionOptions[j].split(" ")[2]
                collectionOptionValues.append(collectionOptionValue)
            else:
                collectionOptionName = collectionOptions[j].split("+")[0]
                collectionOptionNames.append(collectionOptionName)
                collectionOptionValue = collectionOptions[j].split("+")[1]
                collectionOptionValues.append(collectionOptionValue)

        temp3 = itemHTML[i].split("<span class=\"text\">")
        craftItem = []

        # 재료 아이템
        for j in range (1, len(temp3)):
            t_itemName = temp3[j].split("</span>")[0]
            itemName = t_itemName.strip()
            craftItem.append(itemName)

        print("컬렉션 등급: " + collectionLevel + "\n컬렉션 이름: " + collectionName)  
        print("컬렉션 옵션")
        for j in range (len(collectionOptions)):
            #print(collectionOptions[i])
            print(collectionOptionNames[j])
            print(collectionOptionValues[j])
        print("---------")
        print("재료")
        for j in range (len(craftItem)):      
            print(craftItem[j])
        print("---------")

        writeIntoFIleCollection(collectionName, collectionLevel, collectionOptions, collectionOptionNames, collectionOptionValues, craftItem)

    #while True:

        #if temp3[0] == "":
            #break
        #temp3 = temp3[1].split("</span>")[0]
        #temp3 = temp3.strip()
        #print(temp3)

    #print(itemHTML[1])

    #print("등록 아이템: " + temp3)
```

```python
def getFullHTMLFromURL(url):
    fp = urllib.request.urlopen(url)
    mybytes = fp.read()

    mystr = mybytes.decode("utf8")
    fp.close()

    return mystr
```

```python
def writeIntoFIleCollection(collectionName, collectionGrade, collectionOption=[], collectionOptionNames=[], collectionOptionValues=[], collectionIngredient=[]):
    f = open("L2MCollectionList.txt", 'a')
    data = collectionName + "," + collectionGrade + ","
    for i in range(3):
        if len(collectionOption)-1 < i:
            data = data + "NULL,NULL,"
        else:
            data = data + collectionOptionNames[i]+ "," + collectionOptionValues[i]+","
    for i in range(6):
        if len(collectionIngredient)-1 < i and i<5:
            data = data + "NULL,"
        elif len(collectionIngredient)-1 < i and i==5:
            data = data + "NULL"
        elif i<5:
            data = data + collectionIngredient[i] + ","
        else:
            data = data + collectionIngredient[i]
    data = data + "\n"
    f.write(data)
    f.close()
```

```python
def writeIntoFIleEquipment(filename, name, grade, getPlace, _type, destroyPrevent,penetrate, skill, skillDes, optionsName=[], optionsValue=[]):
    f = open(filename, 'a')
    data = name + "," + grade + "," + getPlace +"," + _type +","
    for i in range(7):
        if len(optionsName)-1 < i:
            data = data + "NULL,NULL,"
        else:
            data = data + optionsName[i]+ "," + optionsValue[i]+","
    if destroyPrevent == 1:
        data = data + "TRUE,"
    else:
        data = data + "FALSE,"
    if penetrate == 1:
        data = data + "TRUE,"
    else:
        data = data + "FALSE," 
    data = data + skill + "," + skillDes
    
    data = data + "\n"
    f.write(data)
    f.close()
```
